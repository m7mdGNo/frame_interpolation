```python
import torch
import config
from Generator import Generator
from utils import generate_images,generate_image
import numpy as np
import cv2
```

```python
torch.backends.cudnn.benchmark = True
```

```python
gen = Generator(input_channels=6,output_channels=3, features=64).to(config.DEVICE)

gen.load_state_dict(torch.load('gen.pt'))
```

```python
frames = []
```

```python
cap = cv2.VideoCapture('clibs/arthur.mp4')
while(cap.isOpened()):
    ret, frame = cap.read()
    if not ret:
        break
    frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    frame = cv2.resize(frame,(256,256))
    frames.append(frame)
cap.release()
cv2.destroyAllWindows()
```

```python
def inter_4x(frame1,frame2,gen,out):
    global n
    input1 = np.concatenate((frame1,frame2),axis=2)
    output1 = generate_image(gen,input1)
    # input2 = np.concatenate((frame1,output1),axis=2)
    # output2 = generate_image(gen,input2)
    # input3 = np.concatenate((output1,frame2),axis=2)
    # output3 = generate_image(gen,input3)
    out.write(frame1)
    cv2.imshow('frame',frame1)
    cv2.waitKey(1)
    n=n+1
    # out.write(output2)
    # cv2.imshow('frame',output2)
    # cv2.waitKey(1)
    # n=n+1
    out.write(output1)
    cv2.imshow('frame',output1)
    cv2.waitKey(1)
    n=n+1
    # out.write(output3)
    # cv2.imshow('frame',output3)
    # cv2.waitKey(1)
    # n=n+1
    out.write(frame2)
    cv2.imshow('frame',frame2)
    cv2.waitKey(1)
    n=n+1

```

```python
n = 0
fourcc = cv2.VideoWriter_fourcc(*'DIVX')
out = cv2.VideoWriter('output_smoothed8.mp4', 
                         fourcc,
                         60, (256,256))
for frame in range(len(frames)-1):
    inter_4x(frames[frame],frames[frame+1],gen,out)
    # img = np.concatenate((frames[frame],frames[frame+1]),axis=2)
    # gen_output = generate_image(gen,img)
    # gen_output = cv2.resize(gen_output,(1280,720))
    # frame = cv2.resize(frames[frame],(1280,720))
    # frame = cv2.cvtColor(frame,cv2.COLOR_RGB2BGR)
    # gen_output = cv2.cvtColor(gen_output,cv2.COLOR_RGB2BGR)
    # # cv2.imwrite(f'frames/{n}.jpg',frame)
    # out.write(frame)
    # n+=1
    # # cv2.imwrite(f'frames/{n}.jpg',gen_output)
    # out.write(gen_output)
    # n+=1
    
    print(f'frame{n} finished')
out.release()
cv2.destroyAllWindows()
```
